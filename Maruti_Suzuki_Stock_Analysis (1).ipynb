Calculating the Financial Ratios

```python
# Sample balance sheet and income statement data
balance_sheet = {'Total Assets', 'Total Liabilities', 'Current Assets', 'Current Liabilities'}
income_statement = {'Revenue': [1500, 1700], 'Net Income': [200, 250]}

# Convert to DataFrame
import pandas as pd
balance_df = pd.DataFrame(balance_sheet, index=['2022', '2023'])
income_df = pd.DataFrame(income_statement, index=['2022', '2023'])

# Calculate Debt-to-Equity Ratio
balance_df['Debt-to-Equity'] = balance_df['Total Liabilities'] / (balance_df['Total Assets'] - balance_df['Total Liabilities'])

# Revenue Growth
income_df['Revenue Growth'] = income_df['Revenue'].pct_change()

print(balance_df)
print(income_df)
```

```python
## The Fcf is positive in the year 2020 and 2021 and then negative in the year 2022(maybe recovering from covid) and then positive in the year 2023 which gives a scenario that may be the company is recovering its financial position. Negative capital expenditure indicates that the company is not investing in the long term assets and the reason could be high debt levels or any recovery.The end cash position of the company in the year 2021 was maximum and after that declined and is going up in the year 2023. The compnay cash position seems negligible as it is going up and down in the consecutive years. The is change in the cash in the alternative years and the drastic change the first year and the third year. So the common stock dividend paid is negative in the first three years and nan in the current years indicating that the company is recieving the cash from its shareholders the reason could be checked through the annual report.
```

Cummulative Returns and Drawdowns etc

```python
Ashok= yf.download("ASHOKLEY.BO")
```

```python
ashok=Ashok.Close.to_frame()
```

```python
ashok
```

```python
ashok["d_returns"]=np.log(ashok.div(ashok.shift(1)))
```

```python
ashok
```

```python
ashok.dropna(inplace=True)
```

```python
ashok
```

```python
ashok.d_returns.sum()
```

```python
np.exp(ashok.d_returns.sum())
```

```python
ashok["cummreturns"]=ashok.d_returns.cumsum().apply(np.exp)
```

```python
ashok
```

```python
ashok.cummreturns.plot(figsize=(12,8),title="ashok buy and hold",fontsize=12)
plt.show()
```
